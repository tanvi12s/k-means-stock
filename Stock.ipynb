```python
import pandas as pd
data = pd.read_csv("tds.csv")
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
data_for_clustering = data.drop(columns =['record number', 'Stock Return %'])
data_for_clustering.head()
```

```python
kmeans = KMeans(n_clusters=7, random_state=1729)
data["Cluster"] = kmeans.fit_predict(data_for_clustering)
data.head()
```

```python
pairs = [(264,604), (727,750), (342,840), (660,540)]

for pair in pairs: 
    cluster_pair = data.loc[data['record number'].isin(pair),'Cluster'].tolist()
    
    if cluster_pair[0] == cluster_pair[1]:
         print(f"Pair {pair} belongs to the same cluster.")
    else:
        print(f"Pair {pair} does not belong to the same cluster.")
```
